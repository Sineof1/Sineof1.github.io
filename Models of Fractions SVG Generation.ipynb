```python
import math
import random
import os

#os.mkdir('MoF_svgs')

numerators = []
denominators = []
model_types = []
mtype = ['set', 'area', 'numline']

for n in range(200):
    mod = mtype[random.randint(0, 2)]
    num = random.randint(1, 10)
    den = random.randint(1, 10)
    
    if num / den <= 5 and den / num <= 5:
        if mod == 'set' and num / den > 1:
            continue
        numerators.append(num)
        denominators.append(den)
        model_types.append(mod)

fracs = list(set(zip(numerators, denominators, model_types))) #For each fraction, (numerator, denominator, model)
```

```python
for inx, item in enumerate(fracs):
    svgmarkup = '<svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 450 450" width="100%" height="100%" \
                 preserveAspectRatio="xMidYMid" style="transform:scale(0.95, -0.95);">\n'
    wholes = math.ceil(item[0] / item[1]) #Number of wholes needed to represent the fraction with area model.
    partss = item[1]
    partHeight = (((90 / wholes) / 100) * 450) / partss
    partsCounted = item[0] #Used as a counter variable when shading.
    tickWidth = 410 / (partss * wholes)
    lt1 = item[0] / item[1] < 1
    eq1 = item[0] / item[1] == 1
    gt1 = item[0] / item[1] > 1
    
#Area models
    if item[2] == 'area':
        #First image is of the correct number of wholes divided into denominator parts.
        for b in range(wholes):
            svgmarkup += f'<rect x="0" y="{b * (((100 / wholes) / 100) * 450)}" width="95%" height="{90 / wholes}%" \
                         style="fill: none; stroke: black; stroke-width: 2;"></rect>\n'
            for p in range(partss):
                svgmarkup += f'<rect x="0" y="{(b * (((100 / wholes) / 100) * 450)) + (p * partHeight)}" width="95%"\
                             height="{(90 / wholes) / partss}%" style="fill: none; stroke: lightgray; stroke-width: 1">\
                             </rect>\n'
        if lt1 or eq1:
            svgmarkup += '<text x="225" y="425" text-anchor="middle" style="transform-origin:50% 50%;transform:rotate(90deg) scale(1, -1);">1 whole</text>'
        svgmarkup += '</svg>'
        
        f = open(f'MoF_svgs/models_of_fractions_area_{inx + 1}a.svg', 'w+')
        f.write(svgmarkup)
        f.close()

        svgmarkup = '<svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 450 450" width="95%" height="100%" \
                     preserveAspectRatio="xMidYMid" style="transform:scale(0.95, -0.95);">\n'
        for c in range(wholes):
            svgmarkup += f'<rect x="0" y="{c * (((100 / wholes) / 100) * 450)}" width="95%" height="{90 / wholes}%"\
                         style="fill: none; stroke: black; stroke-width: 2;"></rect>\n'
            #Second image is of the correct shading of numerator parts.
            for q in range(partss):
                if partsCounted > 0: partFill = 'rgba(0, 0, 255, 0.3)'
                else: partFill = 'none'
                svgmarkup += f'<rect x="0" y="{(c * (((100 / wholes) / 100) * 450)) + (q * partHeight)}" width="95%"\
                             height="{(90 / wholes) / partss}%" style="fill: {partFill}; stroke: lightgray;\
                             stroke-width: 1"></rect>\n'
                partsCounted -= 1;
        svgmarkup += '</svg>'
        
        g = open(f'MoF_svgs/models_of_fractions_area_{inx + 1}b.svg', 'w+')
        g.write(svgmarkup)
        g.close()
        break

#Set models
    elif item[2] == 'set':
        circspos = [(225, 225)]
        for d in range(partss - 1):
            circpos = (random.randint(25, 425), random.randint(25, 425))
            #Preventing collisions
            while any([math.sqrt((circspos[a][0] - circpos[0])**2 + (circspos[a][1] - circpos[1])**2) <= 44 for a, val in enumerate(circspos)]):
                circpos = (random.randint(25, 425), random.randint(25, 425))
            circspos.append(circpos)
        for item in circspos:
            svgmarkup += f'<circle cx="{item[0]}" cy="{item[1]}" r="20" style="fill:none;stroke:gray;stroke-width:2;" />\n'
        svgmarkup += '</svg>'
        
        h = open(f'MoF_svgs/models_of_fractions_set_{inx + 1}a.svg', 'w+')
        h.write(svgmarkup)
        h.close()
        
        svgmarkup = '<svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 450 450" width="100%" height="100%" \
                     preserveAspectRatio="xMidYMid" style="transform:scale(0.95, -0.95);">\n'
        
        for item in circspos:
            if partsCounted > 0: partFill = 'rgba(255, 255, 0, 0.3)'
            else: partFill = 'none'
            svgmarkup += f'<circle cx="{item[0]}" cy="{item[1]}" r="20" style="fill:{partFill};stroke:gray;stroke-width:2;" />\n'
            partsCounted -= 1;
        svgmarkup += '</svg>'
        
        j = open(f'MoF_svgs/models_of_fractions_set_{inx + 1}b.svg', 'w+')
        j.write(svgmarkup)
        j.close()
        
#Number line models
    elif item[2] == 'numline':
        svgmarkup = '<svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 450 450" width="100%" height="100%" \
                     preserveAspectRatio="xMidYMid">\n'
        svgmarkup += '<path d="M 5 225 L 10 220 L 5 225 L 10 230 L 5 225 L 445 225 L 440 220 L 445 225 L 440 230 L 445 225 Z" style="fill:none;stroke:gray;" />\
                      <path d="M 20 220 L 20 230 Z" style="fill:none;stroke:gray;" />\n\
                      <path d="M 430 220 L 430 230 Z" style="fill:none;stroke:gray;" />\n'
        for f in range(partss * wholes - 1):
            svgmarkup += f'<path d="M {(f + 1) * tickWidth + 20} 220 L {(f + 1) * tickWidth + 20} 230 Z"\
                         style="fill:none;stroke:gray;" />\n'
        for g in range(wholes + 1):
            svgmarkup += f'<text x="{20 + g * partss * tickWidth}" y="250" text-anchor="middle"\
                         style="font-size:16px;fill:black;stroke:none;">{str(g)}</text>\n'
        svgmarkup += '</svg>'
        
        k = open(f'MoF_svgs/models_of_fractions_numline_{inx + 1}a.svg', 'w+')
        k.write(svgmarkup)
        k.close()
        
        svgmarkup = '<svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 450 450" width="100%" height="100%" \
                     preserveAspectRatio="xMidYMid">\n'
        svgmarkup += '<path d="M 5 225 L 10 220 L 5 225 L 10 230 L 5 225 L 445 225 L 440 220 L 445 225 L 440 230 L 445 225 Z" style="fill:none;stroke:gray;" />\
                      <path d="M 20 220 L 20 230 Z" style="fill:none;stroke:gray;" />\n\
                      <path d="M 430 220 L 430 230 Z" style="fill:none;stroke:gray;" />\n'
        
        for m in range(partss * wholes - 1):
            svgmarkup += f'<path d="M {(m + 1) * tickWidth + 20} 220 L {(m + 1) * tickWidth + 20} 230 Z"\
                         style="fill:none;stroke:gray;" />\n'
        for n in range(wholes + 1):
            svgmarkup += f'<text x="{20 + n * partss * tickWidth}" y="250" text-anchor="middle"\
                         style="font-size:16px;fill:black;stroke:none;">{str(n)}</text>\n'
        svgmarkup += f'<rect x="20" y="220" width="{tickWidth * item[0]}" height="10"\
                     style="stroke:none;fill:rgba(0, 255, 0, 0.3);" />\n'
        svgmarkup += '</svg>'
        
        z = open(f'MoF_svgs/models_of_fractions_numline_{inx + 1}b.svg', 'w+')
        z.write(svgmarkup)
        z.close()
```
